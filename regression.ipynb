```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df1=pd.read_csv("Bengaluru_House_Data.csv")
df1.info()
```

```python
df2=df1.drop(['availability','society','area_type','balcony'],axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3=df2.dropna()
df3.isnull().sum()
df3.shape
```

```python
df3['size'].unique()
```

```python
df3['bhk']=df3['size'].apply(lambda x:int(x.split(' ')[0])) 
df3.head()
```

```python
df3.bhk.unique()
```

```python
df3['total_sqft'].unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

df3[~df3['total_sqft'].apply(is_float)].head(10)
```

```python
def total_sqft_to_float(x):
    tokens=x.split('-')
    if(len(tokens)==2):
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None  
```

```python
total_sqft_to_float('2342')
total_sqft_to_float('3067 - 8156')
```

```python
df4=df3.copy()
df4['total_sqft']=df3['total_sqft'].apply(total_sqft_to_float)
df4.head()
```

```python
df4.loc[188]
```

```python
df4['price_per_sq_feet']=df4['price']*100000/df4['total_sqft']
df4.head()
```

```python
df4.location.unique()
```

```python
# df4.location=df4.location.apply(lambda x: x.strip())
```

```python
location_stats=df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_less_than_tenp=location_stats[location_stats<=10]
location_less_than_tenp
```

```python
df4.location=df4.location.apply(lambda x: 'other' if x in location_less_than_tenp else x) 
df4.head(20)
```

```python
df4.shape
```

```python
df5=df4[~(df4.total_sqft/df4.bhk<400)]
df5.shape
```

```python
df5.price_per_sq_feet.describe()
```

```python
def remove_pps_outliers(df):
    df_out=pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sq_feet)
        st=np.std(subdf.price_per_sq_feet)
        reduced_df=subdf[(subdf.price_per_sq_feet>(m-st))&(subdf.price_per_sq_feet<=(m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out    
```

```python
df6=remove_pps_outliers(df5)
df6.shape
```

```python
df6.bath.unique()
```

```python
df6[df6.bath>=10]
```

```python
df7=df6.drop(['size','price_per_sq_feet'],axis='columns')
df7.head()
```

```python
dummies=pd.get_dummies(df7.location)
dummies=dummies.astype(int)
dummies
```

```python
df8= pd.concat([df7,dummies.drop('other',axis='columns')],axis='columns')
df8.head()
```

```python
df8=df8.drop('location',axis='columns')
df8.shape
```

```python
x=df8.drop('price',axis='columns')
x
```

```python
y=df8['price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
reg.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)

cross_val_score(LinearRegression(),x,y,cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearch(x,y):
    algos = {
        'linear_regression' : {
            'model' : LinearRegression(),
            'params' : {
                'fit_intercept': [True, False],
                'positive': [True, False]
            }
        },
        'lasso':{
            'model': Lasso(),
            'params':{
                'alpha':[1,2],
                'selection':['random','cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params' : {
                'criterion': ['mse','friedman_mse'],
                'splitter' : ['best','random']
            }
        }
    }
    scores=[]
    cv=ShuffleSplit(n_splits=5, test_size=0.2,random_state=0)
    for algo_name, config in algos.items():
        gs=GridSearchCV(config['model'],config['params'], cv=cv, return_train_score=False )
        gs.fit(x,y)
        scores.append({
            'model' : algo_name,
            'best_score' : gs.best_score_,
            'best_params' : gs.best_params_
        })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearch(x,y)
```

```python
def predict_price(location,sqft,bath,bhk):
    loc_index= np.where(x.columns==location)[0][0]

    arr=np.zeros(len(x.columns))
    arr[0]=sqft
    arr[1]=bath
    arr[2]=bhk
    if loc_index>=0:
        arr[loc_index]=1

    return reg.predict([arr])[0] 
    
```
